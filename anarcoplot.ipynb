```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sb.set_style('whitegrid')
```

```python
x1 = np.linspace(1, 3, 1000)
x2 = np.linspace(0, 4, 10)
```

```python
f = 2 + np.sqrt(-(x1-2)**2 + 1)
g = 2 - np.sqrt(-(x1-2)**2 + 1)
h = 3*x2-3
i = -3*x2 + 9
j = 0.2*x2 + 1.7
```

```python
p = plt.plot(x1,f,'r',
             x1,g,'r',
             x2,h,'r',
             x2,i,'r',
             x2,j,'r')

plt.ylim((0, 4))
p[0].axes.set_aspect('equal')
```
